```python
# Import the library
from PIL import Image, ImageOps
import numpy as np
import math
```

```python
# ── Load image ────────────────────────────────────
img = Image.open('cameraman.jpg').convert('L')
```

### 1. Scale (increase size)
Double the image dimensions using `Image.resize()` with high-quality resampling `(LANCZOS)`.

```python
# Get the size of the image

# scaling factors
w, h = img.size
print(f'Original size: {w}x{h}')

sx, sy = 2, 2
```

```python
# ── 1. Increase size (scale ×2) ────────────────────

# Resize the image using PIL's built-in method
scaled = img.resize((w * sx, h * sy), Image.LANCZOS)
```

```python

# Save the scaled image and print the sizes (The new image name should be "task1_1_scaled.jpg")
scaled.save("task1_1_scaled.jpg")
print(f"Original: {img.size} → Scaled: {scaled.size}")
```

 non-uniform scale (cx=2, cy=1) → stretch horizontally only

```python
non_uniform = img.resize((w * 2, h * 1), Image.LANCZOS)
non_uniform.save('task1_1_nonuniform.jpg')
print(f'Non-uniform size: {non_uniform.size}')
```

### 2. Rotate 120°
Rotate the image by 120 degrees, expanding the canvas to fit the full rotated image.

```python
# ── 2. Rotate 120 degrees ──────────────────────────
# Save the scaled image and print the sizes (The new image name should be "task1_2_rotated.jpg")
rotated = img.rotate(120, expand=True)
rotated.save("task1_2_rotated.jpg")
print(f"Rotated size: {rotated.size}")
```

### 3. Shear

```python
# -- c. Get the image dimensions ────────────────────
w, h = img.size
```

```python
# -- d. define the shear matrix ──────────────────────────
# Choose X-axis or Y-axis shear. The shear factor controls how much the image slants — start with 0.5 then experiment.
shx = 0.5               # x-axis shear factor (try 0.1, 0.3, 0.8)
new_w = int(w + abs(shx) * h)
tx = -shx * h if shx > 0 else 0   # offset so image isn't clipped
```

```python
# -- e. Apply the shear transformation to the image ──────────────────────────
# PIL's transform() takes the inverse affine matrix:
# [ 1    shx   tx ]
# [ shy  1     ty ]
sheared = img.transform(
    (new_w, h),
    Image.AFFINE,
    (1, shx, tx, 0, 1, 0),
    resample=Image.BICUBIC,
    fillcolor=0
)
```

```python
# -- f. Save the sheared image(The new image name should be "task1_3_sheared.jpg") 
sheared.save("task1_3_sheared.jpg")
print(f"Sheared size: {sheared.size}")
```

# Intensity Transformations
Negative · Log · Power Law (Gamma)

```python

# ── 1. Negative ──────────────────────────────────── 
# Method 1: NumPy array manipulation
arr = np.array(img)
neg_np = 255 - arr
Image.fromarray(neg_np.astype(np.uint8)).save('task2_1_negative_numpy.jpg')
print('Negative (NumPy) saved')

```

```python
# Method 2: PIL's ImageOps
neg_pil = ImageOps.invert(img)
neg_pil.save('task2_1_negative_pil.jpg')
print('Negative (PIL) saved')
```

```python
# ── 2. Log transformation ──────────────────────────

# s = c · log(1 + r) --> We need to find c.
# We want the maximum output value (s) to be 255 when the maximum input value (r) is 255:
# 255 = c · log(1 + 255)
# c = 255 / log(1 + 255)
arr_f = np.array(img).astype(np.float64)
c = 255 / math.log(1 + 255)


# Apply the log transformation to each pixel
log_img = c * np.log(1 + arr_f)
log_img = np.clip(log_img, 0, 255).astype(np.uint8)
Image.fromarray(log_img).save('task2_2_log.jpg')
print('Log transformation saved')
```

```python

# ── 3. Power-law / Gamma correction ───────────────
gamma = 0.5    # γ < 1 brightens, γ > 1 darkens
c_gamma = 255.0 / (255.0 ** gamma)

gamma_img = c_gamma * (arr_f ** gamma)
gamma_img = np.clip(gamma_img, 0, 255).astype(np.uint8)
Image.fromarray(gamma_img).save('task2_3_gamma.jpg')
print(f'Gamma correction (γ={gamma}) saved')
```
